```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Mengambil data dari komgraf.csv menggunakan var dataset
dataset = pd.read_csv('hujan.csv')

X = np.arange(1,13)
figr = plt.figure(figsize=(10, 3))

#Menentukan koordinat
ax = figr.add_axes([0,0,1,1])

#Membuat title dan label
ax.set_title('Arya Jonathan Trijadi G.231.21.0161')
ax.set_xlabel('Bulan')
ax.set_ylabel('Curah Hujan (mm)')

#Membuat bar chart
ax.bar(X + 0.00, dataset['1981-1990'], color = 'b', width = 0.25)
ax.bar(X + 0.25, dataset['1991-2000'], color = 'r', width = 0.25)
ax.bar(X + 0.50, dataset['2001-2010'], color = 'g', width = 0.25)

ax.legend(['1981-1990', '1991-2000', '2001-2010'], ncol=3, loc='upper center',
          frameon=False)

#Menambahkan garis arah horizontal
for d in [100, 200, 300, 400, 500, 600, 700]:
  if d != 700:
    plt.axhline(y=d, color='gray', linestyle='-', linewidth=0.3)
  else:
    plt.axhline(y=d, color='gray', linestyle=' ', linewidth=0.3)

#Mengatur skala sumbu X dari 1 - 12
ax.set_xticks(X + 0.25)
ax.set_xticklabels(np.arange(1, 13))

plt.show()
```

```python
import datetime
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as matdates

#Membuat data
data_1 = [23, 25.2, 25.5, 24.8, 22.5, 22.2, 23, 22, 25.9]
data_2 = [23.5, 25.3, 25.1, 25.5, 24.9, 25, 24.5, 24, 26]
data_3 = [24, 25, 23.8, 23.9, 24, 24.1, 24, 22.9, 23]

dates = pd.date_range(start='2013-07-24', periods=9, freq='3H')

plt.figure(figsize=(10, 4))

#Membuat plot dengan label yang tertera
plt.plot(dates, data_1, marker='o', color='b', label='Asimilasi')
plt.plot(dates, data_2, marker='o', color='r', label='Observasi')
plt.plot(dates, data_3, marker='o', color='y', label='Non Asimilasi')

#Menambahkan label sumbu X, Y dan judul plot
plt.xlabel('Waktu')
plt.ylabel('Celcius')
plt.title('Suhu Permukaan')

#Menambahkan garis arah horizontal
for d in [20,21, 22, 23, 24, 25, 26, 27]:
  plt.axhline(y=d, color='gray', linestyle='-', linewidth=0.3)

#Mengatur sumbu X sebagai skala waktu
plt.gca().xaxis.set_major_formatter(matdates.DateFormatter('%H:%M UTC'))
plt.gca().xaxis.set_major_locator(matdates.AutoDateLocator())

plt.legend(loc="lower right")
plt.show()
```
